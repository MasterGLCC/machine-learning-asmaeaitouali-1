```python
import numpy as np
# PERCEPTRON FROM SCRATCH 
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]], dtype=float)
 
y = np.array([0, 0, 0, 1], dtype=float)

#  Fonction d'activation 
def sigmoid(z):
    # Transforme n'importe quelle valeur en probabilité (0 → 1)
    return 1 / (1 + np.exp(-z))


#Binary Cross Entropy Loss
def bce_loss(y, y_hat):
    # petit nombre pour éviter log(0)
    eps = 1e-9  
    
    # formule BCE :
    # L = - [ y log(ŷ) + (1-y) log(1-ŷ) ]
    return -np.mean(
        y * np.log(y_hat + eps) +
        (1 - y) * np.log(1 - y_hat + eps)
    )


#Classe Perceptron
class Perceptron:

    # Initialisation du modèle
    def __init__(self, n_inputs, lr=0.5):
        # poids initialisés à 0
        self.w  = np.zeros(n_inputs)
        # biais initialisé à 0
        self.b  = 0.0 
        # learning rate (pas d'apprentissage)
        self.lr = lr                  


    # prédiction
    def forward(self, X):
        # calcul de la somme pondérée
        z = X @ self.w + self.b

        # application de la fonction sigmoid
        y_hat = sigmoid(z)

        return z, y_hat


    # calcul des gradients
    def backward(self, X, y, y_hat):
        n = X.shape[0]  # nombre d'exemples

        # dérivée simplifiée BCE + sigmoid
        dz = y_hat - y  

        # gradient des poids
        dw = (X.T @ dz) / n

        # gradient du biais
        db = np.sum(dz) / n

        return dw, db


    # Mise à jour des paramètres 
    def update(self, dw, db):
        # règle du gradient descent :
        # w = w - lr * dw
        # b = b - lr * db
        self.w -= self.lr * dw
        self.b -= self.lr * db


    # Entraînement du modèle 
    def train(self, X, y, epochs=2000):
        for epoch in range(epochs):

            # 1. forward pass
            z, y_hat = self.forward(X)

            # 2. calcul de la loss
            loss = bce_loss(y, y_hat)

            # 3. backward pass
            dw, db = self.backward(X, y, y_hat)

            # 4. mise à jour des poids
            self.update(dw, db)

            # affichage toutes les 200 itérations
            if (epoch + 1) % 200 == 0:
                print(f"Epoch {epoch+1:5d} | Loss: {loss:.6f}")


#Entraînement
print("=" * 30)
print("  ENTRAÎNEMENT")
print("=" * 30)
p = Perceptron(n_inputs=2, lr=0.5)
p.train(X, y, epochs=2000)
 
```
