```python
import numpy as np
import random

# Environnement simple

# 5 états : 0 → 4
# objectif = atteindre état 4

n_states = 5
n_actions = 2  # 0 = gauche, 1 = droite

# Q-table (initialisée à 0)
Q = np.zeros((n_states, n_actions))

#  Paramètres
alpha = 0.1      # learning rate
gamma = 0.9      # facteur de réduction
epsilon = 0.2    # exploration

episodes = 1000


#  Fonction environnement
def step(state, action):
    # déplacement
    if action == 0:   # gauche
        next_state = max(0, state - 1)
    else:             # droite
        next_state = min(n_states - 1, state + 1)

    # récompense
    if next_state == 4:
        reward = 1   # objectif atteint
    else:
        reward = 0

    return next_state, reward


#  Entraînement Q-Learning
for episode in range(episodes):

    state = 0  # départ

    while state != 4:

        # stratégie epsilon-greedy
        if random.uniform(0, 1) < epsilon:
            action = random.choice([0, 1])  # explorer
        else:
            action = np.argmax(Q[state])    # exploiter

        # interaction avec l’environnement
        next_state, reward = step(state, action)

        # mise à jour Q-table
        Q[state, action] = Q[state, action] + alpha * (
            reward + gamma * np.max(Q[next_state]) - Q[state, action]
        )

        # passer à l’état suivant
        state = next_state


#  Résultat
print("Q-table :\n")
print(Q)


#  Test du modèle
print("\nTest du chemin appris :")

state = 0

while state != 4:
    action = np.argmax(Q[state])
    print(f"Etat {state} -> Action {action}")

    state, _ = step(state, action)

print("Etat final atteint :", state)
```
