```python
import numpy as np
import pandas as pd

#  Charger les données
df = pd.read_csv("data.csv")

# Nettoyage
if 'Unnamed: 32' in df.columns:
    df = df.drop(columns=['Unnamed: 32'])

if 'id' in df.columns:
    df = df.drop(columns=['id'])

# On prend seulement les variables (pas la cible)
X = df.drop(columns=['diagnosis']).values.astype(float)


#  Centrer les données

# On soustrait la moyenne de chaque colonne
# Important pour PCA

X_mean = np.mean(X, axis=0)
X_centered = X - X_mean
```

```python
#  Calcul de la matrice de covariance
# Elle mesure les relations entre variables

cov_matrix = np.cov(X_centered, rowvar=False)


#  Valeurs propres et vecteurs propres
# eigvals = importance des directions
# eigvecs = directions principales

eigvals, eigvecs = np.linalg.eig(cov_matrix)


# Trier les composantes
# On garde les plus importantes

sorted_indices = np.argsort(eigvals)[::-1]

eigvals = eigvals[sorted_indices]
eigvecs = eigvecs[:, sorted_indices]
```

```python
#  Choisir nombre de composantes
k = 2  # on veut 2 dimensions

components = eigvecs[:, :k]


#  Projection des données
# transformation vers nouvel espace

X_reduced = np.dot(X_centered, components)


#  Résultat
print("Dimensions originales :", X.shape)
print("Dimensions réduites :", X_reduced.shape)
```
